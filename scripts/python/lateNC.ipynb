```python
import psycopg2
import pandas as pd
import datetime
```

```python
# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(
    host='10.7.8.59',
    port='5433',
    dbname='fixtransactions',
    user='scalp',
    password='QAtr@de442'
)

# Define the SQL query
query = """
    SELECT a.tag100, AVG(b.tag52 - b.tag60) as avg_difference, count(b.tag52) as counts
    FROM fixmsg a
    JOIN fixmsg b ON a.tag11 = b.tag11
    WHERE a.tag52 >= TIMESTAMP '2023-02-23'
    AND a.tag52 < TIMESTAMP '2023-02-24'
    AND b.tag52 >= TIMESTAMP '2023-02-23'
    AND b.tag52 < TIMESTAMP '2023-02-24'
    AND a.tag35 = 'AB'
    AND b.tag35 = '8'
    AND b.tag39 = '0'
    GROUP BY a.tag100;
"""

# Execute the query and load the results into a Pandas DataFrame
df = pd.read_sql(query, conn)

# Print the first 5 rows of the DataFrame
print(df.head())

# Close the database connection
conn.close()
```
